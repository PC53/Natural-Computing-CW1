## Particle Swarm Optimisation: A simple step-by-step version 

The version will be less useful for producing impressive results

We first initialise parameters, particles, velocities, personal bests and the global best.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=5                                                    # number of particles
D=2                                                    # number of search space dimensions
a_1=1.02                                               # this is meant to be alpha_1 etc.
a_2=1.02
w=0.7
                                                       # we now assume we know the size of the search space
x=np.random.uniform(low=-10,high=10, size=(N,D))       # positions
v=np.random.uniform(low=-0.1,high=0.1, size=(N,D))     # velocities
fp=1.0e10*np.ones(N)                                   # fitness value initialisation for a minimisation problem
p=np.random.uniform(low=-10,high=10, size=(N,D))       # positions of personal bests
fg=1.0e10;                                             # fitness value initialisation for a minimisation problem
g=np.random.uniform(low=-10,high=10, size=D)           # position of global best
```

We could have written something more meaningful for $g$, but this will quickly be be forgotten.

```python
v=w*v+a_1*np.random.rand(D)*(p-x)+a_2*np.random.rand(D)*(g-x)  # these two lines "are" the PSO algorithm
x=x+v

f=x.T[0]**2+x.T[1]**2                                  # this fitness is too simple, try other functions here

for i in range(N):
    if (f[i]<fp[i]):
        fp[i]=f[i]
        p[i]=x[i]
    if (f[i]<fg):
        fg=f[i]
        g=x[i]

xbnd=max(-x.T[0].min(),-p.T[0].min(),x.T[0].max(),p.T[0].max())
ybnd=max(-x.T[1].min(),-p.T[1].min(),x.T[1].max(),p.T[1].max())
[X,Y] = np.meshgrid(np.linspace(-xbnd,xbnd,100),np.linspace(-ybnd,ybnd,100))
Z=X**2+Y**2
plt.contour(X, Y, Z)
plt.scatter(x.T[0],x.T[1], color='blue', marker='o')
plt.scatter(p.T[0],p.T[1], color='green', marker='^')
plt.scatter(g[0],g[1], color='red', marker='o')
#plt.plot(x.T[0]+v.T[0],x.T[1]+v.T[1], color='lime', marker='o')
plt.show()
```
